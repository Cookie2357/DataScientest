```python
""" Voting Classifyer
Is also an ensemble classifier. It is used to combine a heterogeneous set of methods to get the best out of all of them.
Hard vote: Final label is what is predicted most frequently
Soft vote: everybode predicts, average over prediction wins. (Only recommanded if well calibrated)

Stacking: 
Use basic estimators with parameters, evaluate using cross-validation and then build another method on top of it. 
'Meta-learning algorithm'"""
```

```python
from sklearn.ensemble import StackingClassifier, VotingClassifier, RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_validate, KFold
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

import pandas as pd
df = pd.read_csv('source/diabetes.csv')

df.info()
```

```python
target = df.Outcome
data = df.drop(['Outcome'], axis = 1)
data.head()

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=4)
```

```python
clf_nn = KNeighborsClassifier(n_neighbors=3)
clf_rf = RandomForestClassifier(random_state=123)
clf_lr = LogisticRegression(max_iter=1000)

vclf = VotingClassifier([('knn', clf_nn), ('rf', clf_rf), ('lr', clf_lr)], voting='hard')
```

```python
cv3 = KFold(n_splits=3, random_state=111,shuffle=True)
for clf, label in zip([clf_nn, clf_rf, clf_lr, vclf], ['KNN', 'Random Forest', 'Logistic Regression', 'Voting Classifier']):
    scores = cross_validate(clf, X_train, y_train, cv=cv3, scoring=['accuracy','f1'])
    print("[%s]: \n Accuracy: %0.2f (+/- %0.2f)" % (label, scores['test_accuracy'].mean(), scores['test_accuracy'].std()),
          "F1 score: %0.2f (+/- %0.2f)" % (scores['test_f1'].mean(), scores['test_f1'].std()))
```

```python
sclf = StackingClassifier(estimators=[('knn', clf_nn), ('rf', clf_rf), ('lr', clf_lr)], final_estimator=clf_lr)

scores = cross_validate(sclf, X_train, y_train, cv=cv3, scoring=['accuracy', 'f1'])
    
print("[StackingClassifier]: \n Accuracy: %0.2f (+/- %0.2f)\n" % (scores['test_accuracy'].mean(), scores['test_accuracy'].std()),
      "F1 score: %0.2f (+/- %0.2f)" % (scores['test_f1'].mean(), scores['test_f1'].std()))
```

```python
vclf.fit(X_train, y_train)
sclf.fit(X_train, y_train)

print("Acc :", vclf.score(X_test, y_test))
print("Acc :", sclf.score(X_test, y_test))
```
