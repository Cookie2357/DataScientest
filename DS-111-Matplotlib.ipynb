```python
# It is important to include %matplotlib inline to have plots shown automatically

import matplotlib.pyplot as plt

# So far we learned plt.plot() and plt.bar()
# Both share a lot of features. 

plt.title('My title')
plt.xlabel('My xlabel')
plt.ylabel('My yLabel')

plt.axis([0, 10, 0, 40]) # xmin, xmax, ymin, ymax for single line restriction: xlim, ylim

# Other arguments for plot are linewidth, markers, 
plt.plot([0, 2, 4, 8, 16, 32], 'b-',  label = 'Zweier Potenz')
plt.plot([0, 2, 4, 6, 8, 10], 'g--', label = 'gerade Zahlen')

plt.legend(loc = 'best')

# It is also possible to directly draw on a DF. 

""" Bar Plots are another form of plot. They can be combined with line Plots. """
y1 = [10, 20, 30, 40]
plt.bar([6, 7, 8, 9], y1, width = 0.4)
plt.bar([6, 7, 8, 9], [5, 10, 15, 20], width = 0.4, bottom = y1)
plt.bar([x + 0.4 for x in [6, 7, 8, 9]], [5, 10, 15, 20], width = 0.4)

# Question how do I keep all the others correct? I dont I wanna use it such that only those show!
plt.xticks([4, 8], ['X', 'Y'])
plt.yticks([30], ['Drei0']);

```

```python
# To make a continuous plot you take the interval from 0, 2 with steps 0.01 against continuous function
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
plt.plot(t, np.cos(2 * np.pi * t))
plt.ylim([-2, 2])
plt.text(0.8, 1.1, 'local maximum');
```

```python
# We can also annotate points using arrows

plt.grid()
t = np.arange(0, 5, 0.01)
plt.plot(t, np.sin(2 * np.pi * t) * np.exp(-t))
plt.plot(t, np.exp(-t))
plt.plot(t, -np.exp(-t))
plt.annotate('Asymptote: exp(-x)', xy=(1.5, 0.25), xytext=(2, 0.5), arrowprops={'color': 'green'})
plt.annotate('Asymptote: -exp(-x)', xy=(1.5, -0.25), xytext=(2, -0.5), arrowprops={'color': 'red'});

```

```python
## Insert your code here
plt.axis([-5, 2, 0, 25])
plt.grid(linestyle = '--')
plt.annotate('Zoom', xy = (-1.8, 7.5), xytext = (-0.5, 7.5))
t = np.arange(-5, 2, 0.01)

plt.plot(t, t**2, color ='b', label = '$x^2$')
plt.plot(t, np.exp(-t), color = 'g', linestyle = '--', label = 'exp(-x)')

plt.plot([-3, -3, -2, -2, -3], [5, 10, 10, 5, 5], 'r-', alpha = 0.6)

plt.legend(loc = 'upper right')
plt.title('My summary graph')

plt.axes([.55, 0.4, .2, .2])

plt.plot(t,t**2,'b')
plt.plot(t,np.exp(-t),'g--')


plt.axis([-3,-2,5,10])
# But we limit the axes to the coordinates of the square we have drawn.

plt.xticks([])
plt.yticks([]);
```

```python
# We can display a scatter Plot using .scatter

plt.scatter([4, 6, 2, 5, 3, 5, 2, 3, 1, 5, 2], [5, 7, 3, 2, 2, 4, 5, 1, 1, 5, 2], c = 'red', s = 10);

# We can give every node a different color by giving as many colors as datapoints
```

```python
# Histrogram

plt.hist([[1, 2, 3, 3, 3, 4, 5, 6, 7, 8, 9, 9, 9, 10],[2, 3, 3, 3, 4, 5, 5, 5, 7, 8, 8, 9, 9, 10]] \
         , bins = [0, 2, 4, 6, 8, 10], rwidth = 0.8);

"""
We can also specify range which is min(x) and max(x) by default
We can also set bins = 6
"""

plt.ylabel('Frequencies')
plt.xlabel('Sales')
plt.title('Histogram 2 series')
```

```python
# Histrogram

# Other properties are orientation with value 'horizontal'

plt.hist([[1, 2, 3, 3, 3, 4, 5, 6, 7, 8, 9, 9, 9, 10],[2, 3, 3, 3, 4, 5, 5, 5, 7, 8, 8, 9, 9, 10]] \
         , bins = [0, 2, 4, 6, 8, 10], rwidth = 0.8, histtype = 'barstacked');

"""
We can also specify range which is min(x) and max(x) by default
We can also set bins = 6
"""

plt.ylabel('Frequencies')
plt.xlabel('Sales')
plt.title('Histogram 2 series')
```
