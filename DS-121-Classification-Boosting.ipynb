```python
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

import pandas as pd
import numpy as np

df = pd.read_csv('source/letter-recognition.csv')

df.info()
```

```python
data = df.iloc[:, 1:17]
target = df.letter

data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, train_size=0.3)
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)

dtc.score(X_test, y_test)
y_pred = dtc.predict(X_test)

pd.crosstab(y_test, y_pred, rownames = ['actual'], colnames = ['predict'])
```

```python
from sklearn.metrics import classification_report
ac = AdaBoostClassifier(estimator=dtc, n_estimators=400)
ac.fit(X_train, y_train)
ac.score(X_test, y_test)
```

```python
pd.crosstab(y_test, ac.predict(X_test), rownames=['actual'], colnames=['pred'])
```

```python
bc = BaggingClassifier(n_estimators= 500, oob_score = True)
bc.fit(X_train, y_train)

#What does the oob Score tell us?
bc.oob_score_
```

```python
bc.score(X_test, y_test)
y_pred = bc.predict(X_test) 
pd.crosstab(y_test, y_pred, rownames=['actual'], colnames=['predict'])
```
